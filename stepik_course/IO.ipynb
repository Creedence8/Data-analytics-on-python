```python
with open('.\static\dataset_3363_2.txt', 'r') as file:
    stroka = file.readline().strip()

with open('.\static\kek.txt', 'w') as file2:
    letter , digit = '', ''    
    for i in stroka:
        if i.isalpha():
            if letter != '' and digit != '':
                file2.write(letter * int(digit))
                letter , digit = '', ''
            letter = i
        else:
            digit += i
    file2.write(letter * int(digit))
```

```python
with open('.\static\dataset_3363_2.txt', 'r') as inn:
    s = [x.lower() for x in inn.read().split()]

counter = {}
for i in s:
    if i in counter:
        counter[i] += 1
    else:
        counter[i] = 1
        
m = max(counter.values()) 
for k, v in counter.items():
    if v == m:
        print(k, v)
        break
```

```python
with open('.\static\dataset_3363_2.txt', 'r') as inn: 
    s = [[int(z) for z in x.split(';')[1:]] for x in inn.read().strip().split('\n')]
for i in s[:]:
    print(round(sum(i)/3, 9))

for i in range(len(s[0])):
    q = 0
    for j in range(len(s)):
        q+=s[j][i]
    print(round(q/len(s), 9), end=' ')
```
