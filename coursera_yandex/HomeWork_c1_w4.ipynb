```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from math import sqrt
%matplotlib inline
```

```python
chi = stats.chi2(6)
a = chi.rvs(1000)
data = pd.DataFrame(a, columns=['Kde'])
x = np.linspace(0,25, 100)
pdf = chi.pdf(x)
```

```python
plt.hist(a, density=True, label='hist')
plt.plot(x, pdf, label='theoretical pdf')
plt.legend()
plt.show()
```

```python
for n in [3, 10, 30]:
    a = []
    for i in range(1000):
        a.append(np.mean(chi.rvs(n)))
    plt.hist(a, label='n=%s\nx~N(6, %s)' % (n, round(3.464/sqrt(n), 2)), density=True, bins=20)
    x = np.linspace(min(a), max(a), 1000)
    norm = stats.norm(6, 3.464/sqrt(n))
    pdf = norm.pdf(x)
    plt.plot(x, pdf, label='normal')
    plt.legend()
    plt.show()
```
